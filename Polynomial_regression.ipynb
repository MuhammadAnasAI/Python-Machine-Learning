# Polynomial Regression

- Polynomial Regression is the form of regression analysis in which the relationship between the independent and dependent modeled as nth degree polynomial.
- Polynomial Regression is used to model non-linear relationships between variables.


## Mathematics behind the Polynomial Regression:
- Polynomial regression is a form of linear regression where the relationship between the independent variable(s) and the
dependent variable is modeled using a polynomial equation of a certain degree. The degree of the polynomial is the
highest power of the independent variable in the equation.
- In the case of a simple polynomial regression, we have a single independent variable, and the equation takes
the form:
- $$ y = β0 + β1x + β2x^2 + ... + βnx^n $$
- where y is the dependent variable, x is the independent variable, β0 is the intercept, β
is the coefficient of the
independent variable, and n is the degree of the polynomial.
- The coefficients β0, β1, β2, ..., βn are estimated using the method of
least squares, which minimizes the sum of the squared errors between the observed values of y and the
predicted values of y based on the polynomial equation.
- The mathematics behind polynomial regression involves solving a system of linear equations to estimate the coefficients
β0, β1, β2, ..., βn. The system of linear equations is derived
from the least squares method, which is based on minimizing the sum of the squared errors between the observed
values of y and the predicted values of y based on the polynomial equation.
- The system of linear equations can be written as:
$$ \begin{bmatrix}
n & \sum x_i & \sum x_i^2 & ... & \sum x_i
\sum x_iy_i & \sum x_i^2y_i & ... & \sum
\sum x_i^2 & \sum x_i^3 & ... & \sum x_i
\sum x_i^2y_i & \sum x_i^3y_i & ... &
\end{bmatrix} \begin{bmatrix}
\beta_0 \\
\beta_1 \\
\beta_2 \\
\beta_n
\end{bmatrix} = \begin{bmatrix}
\sum y_i \\
\sum x_iy_i \\
\sum x_i^2y_i \\
\end{bmatrix} $$
where n is the number of observations, x_i and y_i are the observed values of the independent
variable and the dependent variable, respectively, and β0, β1, β2, ..., β
are the coefficients to be estimated.

***Import the libraries***

```python
#Import the libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

***Make a dummy dataset***

```python
#Make a dummy dataset for polynomial regression:
np.random.seed(0)
X = 2-3*np.random.normal(0, 1, 20)
y = (X-2)*(X**2) + np.random.normal(-3, 3, 20)
#Reshape the model:
X = X[: , np.newaxis]
y = y[: , np.newaxis]
#Polynomial Transformation:
degree = 2
#Plolynomial features:
poly_features = PolynomialFeatures(degree=2)
```

```python
#Fit the transform:
X_poly = poly_features.fit_transform(X)
#Train the linear model:
model = LinearRegression()
model.fit(X_poly, y)
```

```python
#Predict the values:
y_poly_pred = model.predict(X_poly)
```

```python
import operator
#Plotting the result:
plt.scatter(X, y, color='blue')
sorted_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(X, y_poly_pred),key=sorted_axis)
X, y_poly_pred = zip(*sorted_zip)
plt.plot(X, y_poly_pred, color='magenta')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show() 
```
