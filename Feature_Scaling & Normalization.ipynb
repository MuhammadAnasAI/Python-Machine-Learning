***Standard Scaling using sklearn***
- Range(-1, to +1)

***import the libaries***

```python
#Import the libraries:
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
#Make the dataset:
df = pd.DataFrame({
    'age': [20, 30, 40, 50, 60],
    'height' : [165, 175, 185, 195, 205],
    'weight' : [55, 60, 65, 70, 75]
})
#Check the rows of dataset:
df.head()
```

```python
#Make the StandardScaler:
scaler = StandardScaler()
#Fit the model:
scaled_df = scaler.fit_transform(df)
#Print the scaled_df
scaled_df
```

```python
#Make the Column form:
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
#Print the data:
scaled_df
```

## Transformation of Data
- To change the non normal data into the normal from of data
- To change the data into the form which is easy to understand and easy to process
- To change the data into the form which is easy to store and easy to retrieve


***Import the Libraries***

```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the non Normal dataset:
np.random.seed(0)
df = np.random.exponential(size=100, scale=2)
df = pd.DataFrame(df, columns=['values'])
#Check the Dataset:
df.head()
```

```python
#Make a histogram to check the data:
sns.histplot(df['values'], kde=True)
```

## There are many methods used to transform the non normal data into the normal dataset , but we use some methods to normalize the data 
## These are given below:
- Box-Cox Method
- Yeo-Johnson Method
- Log Transformation
- Square Root Transformation
- Power Transformation
- Quantile Transformer

***Import the Scikit_Learn library sklearn for the Normalization of Data `PowerTransformer`***

```python
#Import the scikit learn libraries for the transformation:
from sklearn.preprocessing import PowerTransformer
```

***Applying the Box Cox Method***
- Box Cox method is applied when the data will be only positive integers

```python
#Apply the Method of Box Cox:
pt_boxcox = PowerTransformer(method='box-cox', standardize=False)
#Fit the model:
df['Box_Cox'] = pt_boxcox.fit_transform(df[['values']] , +1)
#Check the rows of dataset:
df['Box_Cox'].head() 
```

```python
#Make a Histogram plot:
sns.histplot(df['Box_Cox'], kde=True)
```

***Apply the Yeo-Johnson Method***
- It is used for both negative and postive dataset variables

```python
#Apply the method:
pt_yeo_johnson = PowerTransformer(method='yeo-johnson', standardize=False)
#Fit the method:
df['Yeo_Johnson'] = pt_yeo_johnson.fit_transform(df[['values']])
#Check the rows of data:
df.head()
```

```python
#Make a Histplot:
sns.histplot(df['Yeo_Johnson'], kde=True)
```

```python
#Make a histplot with conditions:
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```

***Quantile Transfomation***
- This method is used for good normalization of Dataset
- It's Range is all -ive & +ive values 

```python
#Import the library of quantile transformation:
from sklearn.preprocessing import QuantileTransformer
```

```python
#import Ignore the warning
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
#Apply the method:
qt_normal = QuantileTransformer(output_distribution='normal')
#Fit the model:
df['Quantile'] = qt_normal.fit_transform(df[['values']])
#To check the rows:
df['Quantile'].head()
```

```python
#Make a Histplot of quntile:
sns.histplot(df['Quantile'], kde=True)
```

## L1 and L2 Normalization of Data
- `L1 Normalization:`
  - Also scaled each sample (row) but with a different apporach
  - Ensuring the sum of the Absolute values 1 in each row
  - It is calculated as the sum of absolute vector values
- `L2 Normalization:`
  - It is used for rescale each sample (row) to have unit norm.
  - This used for dealing with textual data
  - Used for NLP purpose
  - It is calculated as the square root of the sum of squared vector values

***L1 Normalization***

```python
#Import the libraries:
from sklearn.preprocessing import Normalizer
```

```python
#Make a Data:
data = [[1, 1, 1], [1, 1, 0], [1, 0, 0]]
df = pd.DataFrame(data)
#Make a histplot:
sns.histplot(data, kde=True)
```

```python
#Apply the model:
normalizer = Normalizer(norm='l1')
#Fit the model:
normalizer = normalizer.fit_transform(data)
#Print the data :
print(normalizer)
```

```python
#Make a histplot:
sns.histplot(normalizer, kde=True)
```

***L2 Normalization***

```python
#Apply the L2 Normalization:
normalizer_2 = Normalizer(norm='l2')
#Fit the model:
normalizer_2 = normalizer_2.fit_transform(data)
#Print the data :
print(normalizer_2)
```

```python
#Make a histplot:
sns.histplot(normalizer_2 , kde=True)
```

## Log Transformation of Data
- It is use for coverting the high integers variables into the sample data parts
 

***Import the Libraries***

```python
#Import libraries:
import pandas as pd
import numpy as np
```

```python
#Apply the data for log transformation:
df = pd.DataFrame({"Values": [1, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 100000]})
#Check the rows:
df.head(13) 
```

```python
#Make a histplot:
sns.histplot(data=df, kde=True)
```

```python
#Fit the log Transformation:
df['log_transform'] = np.log(df['Values'])
#Check the rows:
df.head(13)
```

```python
#Fit the log10 Transformation:
df['log_transform_1'] = np.log10(df['Values'])
#Check the rows:
df.head(13)
```

```python
#Fit the log2 Transformation:
df['log_transform_2'] = np.log2(df['Values'])
#Check the rows:
df.head(13)
```

```python
#Make a histplot:
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```
