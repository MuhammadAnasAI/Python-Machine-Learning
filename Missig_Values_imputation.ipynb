### Types of missing values:
- **Missing Completely At Random (MCAR)**: Missing values are randomly distributed throughout the dataset and
are not related to any specific variable or pattern.
- **Missing At Random (MAR)**: Missing values are related to observed variables, but not to
unobserved variables.
- **Missing Not At Random (MNAR)**: Missing values are related to unobserved variables

### Random Forest Algorithm for imputation of missing values:

```python
#Load the dataset:
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#Check the null value:
df.isnull().sum().sort_values(ascending=False)
```

```python
#Drop the column of deck:
df.drop('deck', axis=1, inplace=True)
```

```python
#Apply the label encoder on categorical columns:
columns_cat = ['sex', 'embarked', 'who', 'class', 'embark_town', 'alive']
le = {}
for col in columns_cat:
    le[col] = LabelEncoder()
    df[col] = le[col].fit_transform(df[col])
    le[col] = le
```

```python
#Check the values:
df.head()
```

```python
#With out missing values columns:
df_with_missval = df[df['age'].isna()]
df_without_missval = df.dropna()
```

```python
#Shape of data:
print(f"Shape of data:",df.shape)
```

```python
print(f"Data with missing values:",df_with_missval.shape)
```

```python
print(f"Data with out missing values:",df_without_missval.shape)
```

```python
#Data head with missing values:
df_with_missval.head()
```

```python
#Data with missing out value:
df_without_missval.head()
```

```python
#Check the columns:
print(df_without_missval.columns)
```

```python
#Split the dataset:
X =     df_without_missval.drop(df[['age']], axis=1)
y =     df_without_missval['age']
```

```python
#Apply the train test split:
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=25, random_state=42)
```

```python
#Apply the Random Forest model:
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
#Fit the model:
rf.fit(X_train, y_train) 
```

```python
#Predict the values:
y_pred = rf.predict(X_test)
```

```python
#Apply the evaluation methods:
print(f"R2 Score:",r2_score(y_test, y_pred))
print(f"MAE: ",np.sqrt(mean_squared_error(y_test, y_pred)))
print(f"MAE :" ,mean_absolute_error(y_pred, y_test))
```

```python
#Missing values prediction:
y_pred = rf.predict(df_with_missval.drop(['age'], axis=1)) 
```

```python
#Ignore the warnings:
import warnings
warnings.filterwarnings("ignore")
```

```python
#Predict the values:
df_with_missval['age'] = y_pred
```
