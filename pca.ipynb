# Principle Component Analysis (PCA) 

- Principle Component Analysis (PCA) is a technique used in `Machine Learning` and `Statistics` to simplify and reduce the dimension of data. The goal of PCA is to identify the patterns in data and find the `corelation` between the `variables`. If `dimensions` are `lines` or `path` that the data can walk along, then a lot of dimensions can be like having two many paths-it make the most important path. PCA can help to classify the simplicity that can find the most important path and getting the read of the less important ones 

## Key terms used in PCA:
- Scaling and features
- Co-variance Matrix
- Eigen Values and Eigen Vectors
- Scree Plot
- Loading Score
- Bi Plot
- Noise Reduction
- Standardization
- Orthogronal Axis

### Principle Component Analysis (PCA) in Python:

**Import the libraries**

```python
#Import the libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Import the sklearn libraries:
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

***Import the dataset***

```python
#load the dataset:
iris = load_iris()
```

***Apply the PCA model***

```python
#Apply the PCA model:
pca = PCA(n_components=2)
```

```python
#Transform the dataset:
iris_pca = pca.fit_transform(iris.data)
```

***A Scatter plot of Dataset with PCA***

```python
#Make a scatter plot of dataset:
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c=iris.target)
plt.xlabel('First PCA')
plt.ylabel('Second PCA')
plt.title('PCA dataset composition')
plt.show()
```

***Check the values of PCA***

```python
#Check the PCA of dataset:
iris_pca
```

***Mean of PCA***

```python
#Mean od iris_pca:
print("Mean of iris_pca: ", iris_pca.mean())
```

## PCA on the tips dataset:

**Import the libraries**

```python
#Import the libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Import the sklearn libraries:
from sklearn.decomposition import PCA
#Import the preprocessing model:
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler 
```

***load the dataset***

```python
#Load the dataset of tips:
df = sns.load_dataset('tips')
#Print the first 5 rows of the dataset
df.head()
```

***Preprocessing of dataset***

```python
#Data Encoding and preprocessing:
le = LabelEncoder()
#Change the categorical data:
cat_features = df.select_dtypes(include='category').columns
#Using the for loop to transform the cat_features:
for feature in cat_features:
    df[feature] = le.fit_transform(df[feature])

#Apply the standard Scaler to the dataset:
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df) 
```

***Apply the PCA***

```python
#Apply the PCA on dataset:
pca = PCA()
#fit the model:
X_pca = pca.fit_transform(df_scaled)
```

***Plot the explained variance ratio***

```python
#Plot the explained variance ratio:
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Corelations')
plt.ylabel('Consective variances')
plt.title('Explained Variances in PCA')
plt.show()
```

**Create a Dataframe to show the loading**
- To check the corelation between datasets. 

```python
#Create a dataframe to show the loading:
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range (pca.n_components_)], index=df.columns)
#Print the Loading:
print(loadings)
```

### Make the Simple Relations:

```python
#Import the libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Import the sklearn libraries:
from sklearn.decomposition import PCA
```

```python
#Load the dataset of tips:
df = sns.load_dataset('tips')
#Print the first 5 rows of the dataset
df.head()
```

```python
#Features of dataset:
X = df[['total_bill', 'tip', 'size']]
#Standardize the dataset:
X = (X-X.mean()) / X.std()
#Perform the PCA:
pca =PCA(n_components=2)
#Fit the model:
pca.fit(X)
```

***Make the Simple relation***

```python
#Transpose the features:
features = pca.components_.T
for i , feature in enumerate(features):
    #Check the simple relation:
    x,y = feature
    color='red' if x>0 and y>0 else 'green' if x<0 and y<0 else 'blue' if x<0 and y<0 else 'purple'
    plt.arrow(0, 0, x, y, color=color, alpha=0.5, width=0.1, head_width=0.3, head_length=0.3, length_includes_head=True)
              #length_include_head=True)
    plt.text(x*1.2, y*1.2, X.columns[i], color=color, ha='center', va='center', fontsize=12)
#Set the axis:
plt.xlim(-1, 1)
plt.ylim(-1, 1)
#Labels of dataset:
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show() 
```
