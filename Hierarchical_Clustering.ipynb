# Hierarchical Clutering


- Hierarchical Clustering is the type of unsupervised learning in which features are divided into the number of groups based
on the same similarity.
- It is a type of clustering algorithm that builds a hierarchy of clusters by merging or splitting existing clusters
- Ploting the data and make a dendrogram plot to link that dataset.


**It is used in the following scenarios:**
- Identifying patterns in data
- Identifying outliers in data
- Identifying clusters in data


## Methods used behind the hierarchical Clustering:
1. **Agglomerative Clustering**: This is a type of hierarchical clustering where we start with
each data point as a separate cluster and then merge them into larger clusters based on
properties.
2. **Divisive Clustering**: In this type of clustering a big group of dataset divided into the subgroups data points.
3. **Group average**: In this method Count the average similarity index of clusters from 5 to 7 in a group.
4. **Single linkage**: In this method the distance between two clusters is the minimum distance between any
two points in the clusters.
5. **Complete linkage**: In this method the distance between two clusters is the maximum distance between any
two points in the clusters.
6. **Ward linkage**: In this method the distance between two clusters is the distance between the
centriod of clusters.


**Import the libraries**

```python
#Import the libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Import the hierarchy from scipy
import scipy.cluster.hierarchy as sch
#Import the AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
#Make the dataset with sklearn:
from sklearn.datasets import make_blobs
```

**Make the labels , Import the Agglomerative Clustering model and fit the model**

```python
X, y = make_blobs(n_samples=100, centers = 5, n_features=5, random_state=42)
#Perform the hierarchical clustering with Agglomative Clustering:
clustering = AgglomerativeClustering(n_clusters=5, linkage='ward')
#Fit the model
clustering.fit(X)
```

**Dendrogram of Dataset**

```python
#Plot the dendrogram of dataset:
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

### Make the Scatter plot and Heatmap plot to check the Hierarchy of dataset:

```python
#Import the libraries:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#Import the hierarchy from scipy
import scipy.cluster.hierarchy as sch
#Import the AgglomerativeClustering
from sklearn.cluster import AgglomerativeClustering
```

***Import the dataset***

```python
#Import the dataset:
df = sns.load_dataset('tips')
df.head()
```

***Make the features and fit the Agglomerative Clustering***

```python
#Make the features of dataset:
feature = df[['total_bill', 'tip']]
#Number of clusters:
n_clusters = 3
#Make the agglomerative cluster of dataset:
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
#Make the agg_label:
agg_labels = agg_clustering.fit_predict(feature)
#Add into the dataset:
df['agg_label'] = agg_labels
#clustering.fit(feature)
```

**A Scatter plot ot check the dataset Clusters**

```python
#Make a scatter plot:
sns.scatterplot(data=df, x='total_bill', y='tip', hue=agg_labels, palette='viridis')
```

***Datasets Dendrogram***

```python
#Plot the dendrogram of dataset:
dendrogram = sch.dendrogram(sch.linkage(feature, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

***To check the co-relation of datasets***

```python
#Make the features of dataset:
feature = df[['total_bill', 'tip', 'size']]
#Plot the heatmap to check the co-relation of dataset:
sns.heatmap(feature.corr(), annot=True)
```

***A Dendrogram of datapoints***

```python
#Make the features of dataset:
feature = df[['total_bill', 'tip', 'size']]
#Compute the linkage matrix:
linkage_matrix = sch.linkage(feature, method='ward')
#Plot the dendrogram of dataset:
dendrogram = sch.dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

**A Heatmap of datapoint to check its relationship**

```python
#Make a heatmap:
sns.heatmap(feature, cmap='coolwarm', annot=True)
```

**Transpose of datasets**

```python
#Transpose of features:
feature_T = feature.T
#Compute the linkage matrix:
linkage_matrix = sch.linkage(feature_T, method='ward')
#Plot the dendrogram of dataset:
dendrogram = sch.dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```
