## Methods use for Encoding and Inverse Transfromation
The encoding and inverse transformation methods used in this study are as follows:
- **StandardScaler**: This method is used to scale the data to have zero mean and unit variance
- **MinMaxScaler**: This method is used to scale the data to a specified range
- **OneHotEncoder**: This method is used to convert categorical variables into numerical variables
- **LabelEncoder**: This method is used to convert categorical variables into numerical variables
- **OrdinalEncoder**: This method is used to convert categorical variables into numerical variables

### Dealing the Missing values with Mean, Median and Mode:
In this section, we will discuss how to deal with missing values in a dataset using mean, median
and mode. We will use the pandas library to perform these operations.

```python
#Handling missing values using mean, median and mode:
#df['age'].fillna(df['age'].median(), inplace=True)
df.fillna({'age': df['age'].median()}, inplace=True)
df.fillna({'embarked':df['embarked'].mode()[0]}, inplace=True)
df.fillna({'embark_town':df['embark_town'].mode()[0]}, inplace=True)
```

```python
#Drop the column of deck:
df.drop('deck', axis=1, inplace=True)
```

```python
#Check the missing value:
df.isnull().sum().sort_values(ascending=False)
```

### Handling the missing value using LabelEncoder and Inverse all these missing values:
- Label Encoders Import the all categorical variable and object variables.
- Inverse Transform: Use inverse transform to convert the encoded values back to the original values.

***Import the Label Encoder***

```python
#Import the label encoder:
le = LabelEncoder()
#Encode the labels
df['sex'] = le.fit_transform(df['sex'])
```

***Lable Encoder***

```python
#Import the dataset:
df = sns.load_dataset('titanic')
```

```python
#check the data:
df.head()
```

```python
#Handling missing values using mean, median and mode:
#df['age'].fillna(df['age'].median(), inplace=True)
df.fillna({'age': df['age'].median()}, inplace=True)
df.fillna({'embarked':df['embarked'].mode()[0]}, inplace=True)
df.fillna({'embark_town':df['embark_town'].mode()[0]}, inplace=True)
```

```python
#Drop the deck column:
df.drop('deck', axis=1, inplace=True)
```

```python
df.head()
```

***Encode the data***

```python
#Change the categorical into  the object:
he_sex = LabelEncoder()
df['sex'] = he_sex.fit_transform(df['sex'])
he_embarked = LabelEncoder()
df['embarked'] = he_embarked.fit_transform(df['embarked'])
he_class = LabelEncoder()
df['class'] = he_class.fit_transform(df['class'])
he_who = LabelEncoder()
df['who'] = he_who.fit_transform(df['who'])
df_embark_town = LabelEncoder()
df['embark_town'] = df_embark_town.fit_transform(df['embark_town'])
                                                     
```

```python
df.head()
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype.name == 'category':
        df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df.info()
```

***Decode the data***

```python
#Revese the object values:
df['sex'] = he_sex.inverse_transform(df['sex'])
df['embarked'] = he_embarked.inverse_transform(df['embarked'])
df['embark_town'] = he_embarked.inverse_transform(df['embark_town'])
df['class'] = he_class.inverse_transform(df['class'])
df['who'] = he_who.inverse_transform(df['who'])
```
